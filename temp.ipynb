```python
import matplotlib.pyplot as plt
from sensor_fusion import get_mapped_points, get_sparse_depthmap
import numpy as np
#from vis_utils import vis_sparse_dmap

bin_path = 'test_data/0000000000.bin'
img_path = 'test_data/0000000000.png'
    
new_velo, new_cam = get_mapped_points(bin_path, img_path)
s_dmap= get_sparse_depthmap(new_cam)
```

```python
s_dmap= get_sparse_depthmap(new_cam)
def vis_sparse_dmap(s_dmap):
    where_0 = np.where(s_dmap == 0)
    s_dmap = s_dmap / np.max(s_dmap)
    s_dmap[where_0] = 255 

    return s_dmap
vis_s_dmap = vis_sparse_dmap(s_dmap)
plt.imshow(vis_s_dmap, cmap='binary')
```
